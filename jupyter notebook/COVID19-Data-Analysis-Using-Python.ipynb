#Matplotlib

```python
fig=plt.figure(figsize=(10,10),dpi=300)
axes=fig.add_axes([0,0,1,1])
axes.bar(df['Country/Region'],df['Confirmed'])
axes.set_title("Confirmed")
axes.set_xlabel("Country/Region")
axes.set_ylabel("Confirmed")
plt.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=df['Country/Region'],y=df['Confirmed']))
fig.update_layout(title='Country/Region',xaxis=dict(title='Country/Region'),yaxis=dict(title='Confirmed'))
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=df['Country/Region'],y=df['Recovered']))
fig.update_layout(title='Country/Region',xaxis=dict(title='Country/Region'),yaxis=dict(title='Recovered'))
```
